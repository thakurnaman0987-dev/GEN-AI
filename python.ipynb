```python
def sum_number(*args):
  sum = 0
  for num in args:
    sum+=num
    return sum
```

```python
#keyword variable lenght argument
def test(**kwargs):
  print(kwargs)
  print(len(kwargs))
  print(type(kwargs))
```

```python
def pass_student(**kwargs):
  pass_student_list = []
  for name, marks in kwargs.items():
    if marks>35:
      pass_student_list.append(name)
      return pass_student_list
```

```python
pass_student(A =10,B = 20,C =40, D =50, E = 60, F = 30)
```
